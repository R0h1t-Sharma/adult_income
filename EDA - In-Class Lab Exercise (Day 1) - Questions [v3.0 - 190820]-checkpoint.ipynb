### About the dataset (Adult Income Data)

The dataset contains information about the individuals' age, gender, and several attributes affecting the income of a person. <br> Attribute information:

**age:** Age of an individual

**workclass:** Working class of the individual

**fnlwgt:** Weights based on the demographic characteristics

**education:** Highest education of the individual

**education.num:** Education encoded with unique numbers

**marital.status:** Marital status of the individual

**occupation:** Occupation of the individual

**relationship:** Relation of the individual in the family

**race:** Race of the individual

**sex:** Gender of the individual (0 = Male, 1 = Female)

**capital.gain:** Gain amount of the individual

**capital.loss:** Loss amount of the individual

**hours.per.week:** Working hours per week

**native.country:** The native country of the individual

**income:** Income of the individual

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the function to plot categories and find the most frequent one
def plot_categories_and_find_max(df, column_name):
    # Plot category frequencies
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=column_name, order=df[column_name].value_counts().index)
    plt.xticks(rotation=90)
    plt.title(f'Category Frequencies for {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.show()
    
    # Find and return the most frequent category
    return df[column_name].mode()[0]

# Load the dataset
#df = pd.read_csv(r"H:\My Drive\EDA +GL\DAY 1\EDA - In-Class Lab Exercise(Day1)\EDA - In-Class Lab Exercise(Day1)\adult_income.csv")

# Replace '?' with NaN and convert columns to appropriate data types
df.replace('?', pd.NA, inplace=True)
df = df.apply(pd.to_numeric, errors='ignore')

# Define the column to analyze
column_to_analyze = 'native.country'

# Call the function and print the result
most_frequent_category = plot_categories_and_find_max(df, column_to_analyze)
print(f"The category with the highest frequency in '{column_to_analyze}' is: {most_frequent_category}")
```

```python
import scipy.stats as stats
```

```python
# type your code here
# Function to check normality
def check_normality(df, columns):
    for col in columns:
        # Q-Q plot
        plt.figure(figsize=(8, 4))
        stats.probplot(df[col].dropna(), dist="norm", plot=plt)
        plt.title(f'Q-Q Plot of {col}')
        plt.show()
        
        # Shapiro-Wilk test
        stat, p = stats.shapiro(df[col].dropna())
        print(f'{col}: Shapiro-Wilk Test p-value = {p}')
        if p > 0.05:
            print(f'{col} appears to be normally distributed.')
        else:
            print(f'{col} does not appear to be normally distributed.')

# Check normality for numerical columns
check_normality(df, numerical_columns)
```

```python
# type your code here
# Calculate skewness
skewness = df[numerical_columns].apply(lambda x: skew(x.dropna()))

# Print skewness values
print("Skewness of numerical features:")
print(skewness)
```

```python
# type your code here
# Calculate kurtosis
kurtosis_values = df[numerical_columns].apply(lambda x: kurtosis(x.dropna()))

# Print kurtosis values
print("Kurtosis of numerical features:")
print(kurtosis_values)
```

```python
# Interpretation
for col, value in kurtosis_values.items():
    if value > 0:
        print(f"{col}: Positive kurtosis ({value:.2f}) indicates a distribution with heavy tails.")
    elif value < 0:
        print(f"{col}: Negative kurtosis ({value:.2f}) indicates a distribution with light tails.")
    else:
        print(f"{col}: Kurtosis is around 0 ({value:.2f}), similar to a normal distribution.")
```

```python
# Filter numerical columns only
numerical_columns = df.select_dtypes(include=[np.number]).columns
correlation_matrix = df[numerical_columns].corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Display the upper triangular correlation matrix
upper_triangular_matrix = correlation_matrix.where(mask)

# Print the upper triangular matrix
print("Upper Triangular Correlation Matrix:")
print(upper_triangular_matrix)
```

```python
# Filter numerical columns only
numerical_columns = df.select_dtypes(include=[np.number]).columns
correlation_matrix = df[numerical_columns].corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Draw the heatmap with the mask
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', center=0, linewidths=0.5, fmt='.2f', vmin=-1, vmax=1)

plt.title('Upper Triangular Correlation Matrix')
plt.show()
```
